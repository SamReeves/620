### Assignment:  Make a simple node graph like the one given.

A graph, in this sense, is a set of nodes and a set of edges.  The nodes represent units, and the edges represent their connections.

```python
import networkx as nx
import matplotlib.pyplot as plt
import warnings
```

```python
G = nx.Graph()

nodes = {0: "Alec",
     1: "Brian",
     2: "Charlie",
     3: "Derek",
     4: "Eric",
     5: "Frankie",
     6: "George",
     7: "Henry",
     8: "Ivan",
     9: "Jake"}

edges = [(0,1), (0,2), (0,3), (0,4),
        (1,2), (1,3), (1,4),
        (2,3), (2,4),
        (3,4),
        (4,5),
        (5,6),
        (6,7), (6,8), (6,9),
        (7,8), (7,9),
        (8,9)]
```

```python
G.add_edges_from(edges)
G = nx.relabel_nodes(G, nodes)
```

```python
warnings.filterwarnings('ignore')
nx.draw(G, with_labels=True)
```
